```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
haar_cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
haar_cascade_face
```

```python
def to_gray(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

### Example 1

```python
#reading the image
test_image = cv2.imread('people.jpg')

#Converting to grayscale
test_image_gray = to_gray(test_image)

# Displaying the grayscale image
plt.imshow(test_image_gray, cmap='gray')
```

```python
#Displaying without cmap command
plt.imshow(test_image_gray)
```

```python
faces_rects = haar_cascade_face.detectMultiScale(test_image_gray, scaleFactor = 1.2, minNeighbors = 5)
print('Faces found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
     cv2.rectangle(test_image, (x, y), (x+w, y+h), (0, 255, 0), 2)
        
plt.imshow(convertToRGB(test_image))
```

### Example 2

```python
family = cv2.imread('family.jpg')

family_gray_img = to_gray(family)

plt.imshow(family_gray_img, cmap='gray')
```

```python
faces = haar_cascade_face.detectMultiScale(family_gray_img, scaleFactor = 1.2, minNeighbors = 1)
print('Faces found: ', len(faces))
```

```python
for (x,y,w,h) in faces:
     cv2.rectangle(family, (x, y), (x+w, y+h), (0, 255, 0), 2)
        
plt.imshow(convertToRGB(family))
```
